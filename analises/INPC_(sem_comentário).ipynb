```r
precos = ipeadata("PRECOS_INPCBR") %>%
  mutate(date = year(date)) %>%
  as_tsibble(index = date)
```

```r
# time plot
precos %>% autoplot()
```

------------------------------------------------

```r
# Taxa de variaçã (dez 1993 = 100) - mensal
precos12 = ipeadata("PRECOS12_INPC12") %>%
  mutate(date = yearmonth(date)) %>%
  as_tsibble(index = date)
```

```r
# time plot
precos12 %>% autoplot()
```

```r
# Seasonal plot
# não identificamos uma sazonalidade
precos12 %>% gg_season(value, labels = "both")
```

```r
# Sub series plot
precos12 %>% gg_subseries(value)
```

```r
# extrair tendencia
precos12 %>% model(STL(value)) %>%
  components() %>%
  autoplot()
```

```r
# ACF plot
precos12 %>% ACF(value, lag_max = 50) %>%
  autoplot()
```

```r
# Lag plot
precos12 %>% gg_lag(value, geom = "point", period = "month")
```
